### **Solución 4**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

#semilla
np.random.seed(856273)

#media teorica
media = (856273 % 10) + 2

#desviación estandar
desviacion = 2

#tamaño de la muestra
n = 30

#generar la distribución
X = np.random.normal(media, desviacion, n)

# Percentiles principales a resaltar (Azules en la imagen)
percentiles_claves = [10, 25, 50, 75, 90]
val_percentiles = np.percentile(X, percentiles_claves)

# Deciles (D1 a D9, es decir, del 10% al 90% de 10 en 10)
deciles = np.arange(10, 100, 10)
val_deciles = np.percentile(X, deciles)

# Cuartiles (Q1=25%, Q2=50%, Q3=75%)
cuartiles = [25, 50, 75]
val_cuartiles = np.percentile(X, cuartiles)

# Imprimir resultados en consola para verificar
print("--- Medidas de Posición ---")
print(f"Cuartiles (Q1, Q2, Q3): {val_cuartiles}")
print(f"Deciles (D1 a D9): {val_deciles}")

# 2. Visualización
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Histograma con la curva de densidad (KDE)
# Nota: La imagen de referencia parece tener una muestra (n) mucho mayor.
# Con n=30 se verá un poco menos "suave", lo cual es normal.
sns.histplot(X, bins=10, kde=True, color='lightblue', alpha=0.5)

# Trazar Deciles (Líneas grises punteadas)
for i, (perc, val) in enumerate(zip(deciles, val_deciles)):
    # Solo dibujamos en gris los que NO son los percentiles principales azules
    if perc not in percentiles_claves:
        plt.axvline(val, color='gray', linestyle='--', alpha=0.4)
        # Etiqueta del decil (D2, D3, D4, etc.)
        plt.text(val, plt.ylim()[1] * 0.4, f'D{i+1}', color='gray',
                 rotation=90, ha='right', va='center', fontsize=9)

# Trazar Percentiles/Cuartiles principales (Líneas azules sólidas)
etiquetas_azules = ['P10\n(D1)', 'P25\n(Q1)', 'P50 (Mediana)\n(D5, Q2)', 'P75\n(Q3)', 'P90\n(D9)']
for val, etiqueta in zip(val_percentiles, etiquetas_azules):
    plt.axvline(val, color='blue', linestyle='-', linewidth=1.8)
    # Etiqueta del percentil
    plt.text(val, plt.ylim()[1] * 0.85, etiqueta, color='blue',
             rotation=90, ha='right', va='center', fontsize=9)

plt.title("Percentiles, Deciles y Cuartiles sobre la Distribución", fontsize=14, fontweight='bold')
plt.xlabel("Valor de X")
plt.ylabel("Frecuencia")

texto_caja = (
    "Percentiles: posición relativa (P10, P25, P50, P75, P90)\n"
    "Deciles: dividen en 10 partes (D1 a D9)\n"
    "Cuartiles: Q1 = 25%, Q2 = 50%, Q3 = 75%\n"
    "Analizar dispersión, concentración y desigualdad"
)

props = dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='gray')
plt.text(0.98, 0.95, texto_caja, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', horizontalalignment='right', bbox=props)

plt.tight_layout()
plt.show()
```

2. Genere dos variables aleatorias (ejemplo: `ingresos` y `gastos`) con distribución normal, usando esa base como **semilla reproducible**.

### **Solución 5**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# semilla
semilla = 856273
np.random.seed(semilla)

# tamaño de muestra
n = 100

#variable X con media de 3000, desviación de 500
ingresos = np.random.normal(loc=3000, scale=500, size=n)

# Variable Y
# sean aproximadamente el 70% de los ingresos, más un "ruido" aleatorio normal
ruido = np.random.normal(loc=0, scale=300, size=n)
gastos = 0.7 * ingresos + ruido

#coeficientes de correlación
#Pearson evalúa la relación lineal
pearson_corr, pearson_p = stats.pearsonr(ingresos, gastos)
#Spearman evalúa la relación monótona (rango)
spearman_corr, spearman_p = stats.spearmanr(ingresos, gastos)

print("--- Resultados del Análisis Bivariado ---")
print(f"Coeficiente de Pearson: {pearson_corr:.4f}")
print(f"Coeficiente de Spearman: {spearman_corr:.4f}")

#Graficar las variables
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

sns.regplot(x=ingresos, y=gastos,
            scatter_kws={'alpha': 0.6, 'color': 'dodgerblue'},
            line_kws={'color': 'red', 'linewidth': 2})

plt.title('Análisis Bivariado: Relación entre Ingresos y Gastos', fontsize=14, fontweight='bold')
plt.xlabel('Ingresos (Variable X)')
plt.ylabel('Gastos (Variable Y)')

# Añadir un recuadro con los resultados de correlación en la gráfica
texto_stats = (
    f"Correlación de Pearson (r): {pearson_corr:.2f}\n"
    f"Correlación de Spearman (ρ): {spearman_corr:.2f}"
)
props = dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='gray')
plt.text(0.05, 0.95, texto_stats, transform=plt.gca().transAxes, fontsize=11,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()
```

**Interpretación:**
* Lo primero que se observa es una correlación positiva, uno crece, el otro también.
* La correlación de Pearson de 0.81 me dice que hay una alta correlación LINEAL positiva (a medida que los ingresos incrementan, los gastos también incrementan de manera proporcional).
* La correlación de Spearman de 0.80, en la cual medimos la monotonía (como lo entiendo, es que lo único que nos importa es que, si una variable sube, la otra también). Entonces, dado el valor, también se puede confirmar aún más el crecimiento consistente en los datos.

Entonces existe una dependencia positiva en nuestras dos variables.

### <span style="color:#2F749F;"><strong>Ejercicio 6: Chi-Cuadrado</strong></span>

En esta parte, analiza si existe **asociación estadística** entre dos variables categóricas simuladas a partir de tu número de cédula.

1. Usa los **6 últimos dígitos** de tu cédula como base para generar una semilla reproducible.
2. Simula una muestra de **60 personas** con dos variables categóricas:
   - `Nivel educativo`: Básico, Medio, Superior.
   - `Acceso a internet`: Sí, No.
3. Crea una **tabla de contingencia** entre ambas variables.
4. Aplica la **prueba de Chi cuadrado de independencia**.
5. Visualiza la distribución conjunta con un mapa de calor (ver la imagen de con la información de ejemplo).
6. Interpreta el resultado.

image.png

### **Solución 6**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Semilla reproducible
semilla = 856273
np.random.seed(semilla)

# Tamaño de la muestra
n = 60

# 2. Simular las nuevas variables categóricas
categorias_sector = ['Urbano', 'Rural', 'Periferia']
categorias_transporte = ['Público', 'Privado']

# Generamos las opciones aleatorias
sector_residencia = np.random.choice(categorias_sector, size=n)
preferencia_transporte = np.random.choice(categorias_transporte, size=n)

# Creamos el DataFrame
df = pd.DataFrame({
    'Sector': sector_residencia,
    'Transporte': preferencia_transporte
})

# 3. Crear la tabla de contingencia
tabla_contingencia = pd.crosstab(df['Sector'], df['Transporte'])

# Reordenamos para que se vea organizado
tabla_contingencia = tabla_contingencia.reindex(index=['Urbano', 'Rural', 'Periferia'], columns=['Público', 'Privado'])
print("Tabla de Contingencia: Sector vs. Transporte")
print(tabla_contingencia)
print("-" * 40)

# 4. Prueba de Chi-cuadrado de independencia
chi2, p_valor, grados_libertad, frecuencias_esperadas = stats.chi2_contingency(tabla_contingencia)

print(f"Resultado Estadístico:")
print(f"Chi² = {chi2:.2f}")
print(f"p-valor = {p_valor:.3f}")
print(f"Grados de libertad = {grados_libertad}")
print("\nFrecuencias esperadas (Teóricas):")
df_esperadas = pd.DataFrame(frecuencias_esperadas,
                            index=tabla_contingencia.index,
                            columns=tabla_contingencia.columns)
print(df_esperadas.round(1))

# 5. Visualizar con un mapa de calor (Heatmap)
sns.set_theme(style="white")
plt.figure(figsize=(8, 5))

# Cambiamos cmap='YlGn' (Yellow to Green) para un estilo diferente
sns.heatmap(tabla_contingencia, annot=True, fmt='d', cmap='YlGn',
            cbar_kws={'label': 'Número de Personas'})

plt.title('Distribución Conjunta: Sector de Residencia vs. Preferencia de Transporte', fontsize=13, pad=15)
plt.ylabel('Sector de Residencia')
plt.xlabel('Preferencia de Transporte')

plt.tight_layout()
plt.show()
```

**Interpretación:**
* De la tabla de contingencia y gráfico de calor puedo decir que en Urbano está equilibrado entre **público** y **privado**. En Rural hay una diferencia más significativa, siendo el transporte **privado** más preferido. Y en las periferias se opta mucho más por el transporte **privado**.
* Entonces, como hacemos la prueba de independencia con ji cuadrado, se manejan las dos hipótesis. **H0** (Las variables son independientes, NO hay asociación entre el nivel educativo y tener internet), **H1** (Las variables están asociadas, el nivel educativo influye o está relacionado con tener internet).
* Como el valor P = 0.352 es mayor al nivel de significancia (0.05), **no existe suficiente evidencia para rechazar** nuestra hipótesis nula, por lo tanto, no podemos decir que el sector esté asociado al tipo de transporte.
* Del valor que me arroja chi cuadrado (2.09) puedo deducir que este es pequeño y, por consiguiente, los números observados no son muy diferentes de los que se esperarían si todo fuera al azar.

## <span style="color:black;"><strong>Parte 3: Análisis  Multivariado</strong></span>

### **Analisis multivariado por Componentes Principales (PCA)**
**Breast Cancer Wisconsin**

El Contexto: ¿Qué es la base de datos "Breast Cancer Wisconsin"?

Un médico extrae una pequeña muestra de tejido de un bulto en el pecho de una paciente (esto se llama biopsia). Esa muestra se mira bajo un microscopio, y un programa de computadora mide 30 características diferentes de las células que aparecen ahí.

**¿Qué miden?** Cosas como el radio de la célula, su textura, su área, qué tan suave o rugoso es su borde, etc.

**El objetivo:** Saber si ese grupo de células es un tumor Benigno (no es cáncer) o Maligno (cáncer).

image.png

#### **Carga del DataSet, estandarización y PCA**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

#el dataframe
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df['diagnosis'] = pd.Categorical.from_codes(cancer.target, cancer.target_names)
df.head()

#Estandarización de variables númericas
scaler = StandardScaler()
X_std = scaler.fit_transform(df[cancer.feature_names])

#el PCA
pca = PCA()
X_pca = pca.fit_transform(X_std)

#visualización del df
df.head()
```

**Observación:**

Lo que entiendo de **estandarización** es básicamente poner todo en una misma escala (por ejemplo, unas características tienen valores muy pequeñitos, otras muy grandes), que no pase que a la hora de emplear el PCA este piense que tal característica es más importante que otra solo por su gran diferencia de valores.

#### **Mapa de Calor (Autovectores)**

```python
# Matriz de componentes (primeros 5)
componentes = pd.DataFrame(
    pca.components_.T[:, :5],
    columns=[f'PC{i+1}' for i in range(5)],
    index=cancer.feature_names
)

# Visualización de autovectores (Top 5 componentes)
sns.heatmap(componentes, annot=False, cmap='coolwarm', center=0)
plt.title('Autovectores (Primeros 5 Componentes)')
plt.ylabel('Variables originales (30)')
plt.xlabel('Componentes principales')
plt.tight_layout()
plt.show()


for i in range(5):
    pesos = pca.components_[i]
    ecuacion = " + ".join([f"{round(pesos[j], 3)}·'{cancer.feature_names[j]}'" for j in range(3)])
    print(f"PC{i+1} = {ecuacion} + ... (27 variables más)")
```

**Observación:**

Entonces, aquí cogemos las 30 características, las licuamos, por decirlo así, y de esto sacamos 5 variables nuevas llamadas componentes principales PC 1, 2, 3, 4 y 5. Entonces, en el PC1 se tendrían combinadas las características que capturan una mayor diferencia entre sí. Es como si necesitara que los pacientes malignos se vayan para un lado y los benignos para el otro. ¿Qué ayuda más a distinguir?


#### **La Varianza Explicada**

```python
# Varianza explicada
varianza = pca.explained_variance_
PVE = pca.explained_variance_ratio_
PVE_acum = np.cumsum(PVE)
componentes_nombres = [f'PC{i+1}' for i in range(len(PVE))]

# DataFrame para visualización
df_varianza = pd.DataFrame({
    'Componente': componentes_nombres,
    'Autovalor': varianza,
    'PVE': PVE,
    'PVE acumulada': PVE_acum})

print("Matriz resumen de varianza explicada (Primeros 5):\n")
print(df_varianza.head().round(3))

# --------------------------
# Gráfico 1: Proporción de varianza explicada
# --------------------------
plt.figure(figsize=(10, 5))
# Solo graficamos los primeros 15 componentes para que se vea claro
sns.barplot(x='Componente', y='PVE', data=df_varianza.head(15), color='skyblue', label='PVE individual')
sns.lineplot(x='Componente', y='PVE acumulada', data=df_varianza.head(15), marker='o', color='darkblue', label='PVE acumulada')

plt.axhline(0.7, color='red', linestyle='--', label='Umbral 70%')
plt.axhline(0.9, color='green', linestyle='--', label='Umbral 90%')

plt.title("Proporción de varianza explicada (Top 15 PCs)", fontsize=15, fontweight="bold", backgroundcolor="#fff9db", pad=15)
plt.ylabel("Proporción de varianza")
plt.xlabel("Componentes principales")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Observación:**

Aquí lo que observo es que, con **solo 4 componentes**, puedo retener más del  80% de la info. En comparación con el ejemplo hecho en clase, que solo tenía 4 características, aquí tenemos 30, esto me ayuda más a dimensionar lo que se quiere lograr hacer con el PCA.

**El Gráfico de Dispersión**

```python
# Guardamos todos los componentes en un DF
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(len(PVE))])

# Agregamos la variable categórica para colorear
df_pca['diagnosis'] = df['diagnosis']

# Visualización del gráfico de dispersión (Scatter plot)
plt.figure(figsize=(8, 6))
# Usamos 'malignant' y 'benign' que son las clases de este dataset
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='diagnosis', palette={'malignant': '#e74c3c', 'benign': '#2ecc71'}, alpha=0.7)

plt.title('Proyección de los datos de Cáncer de Mama en PC1 y PC2')
plt.xlabel(f'PC1 ({PVE[0]*100:.1f}% varianza)')
plt.ylabel(f'PC2 ({PVE[1]*100:.1f}% varianza)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

**Observación:**

Algo que me ayuda a entender el concepto de PCA y la importancia del concepto de ortogonalidad en su práctica, profe, es la analogía de exprimir una naranja. El PC1 la exprime en la dirección donde encuentre más juguito (más varianza), eL PC2, por otro lado, no quiere ser **redundante** y selecciona exprimir en una dirección donde el PC1 no lo haya hecho, totalmente **ortogonal** a PC1. Todo esto para lograr lo que se observa en la gráfica de PC1 VS PC2, casi que se logran identificar por completo los diagnosticos de los pacientes.
